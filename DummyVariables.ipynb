```python
import pandas as pd
```

```python
df=pd.read_csv('homeprices (2).csv')
```

```python
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
combining = pd.concat([df,dummies])
combining
```

```python
final= combining.drop(['town'],axis='columns')
final.head()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x=final.drop('price',axis='columns')
x.head()
```

```python
y=final.price
y.head()
```

```python
x.dropna(inplace=True)
y.dropna(inplace=True)
```

```python
model.fit(x,y)
```

```python
#price in robinsville place for my required area can be predicted like this
model.predict([[2800,0,1,0]])
```

```python
model.score(x,y)
```
